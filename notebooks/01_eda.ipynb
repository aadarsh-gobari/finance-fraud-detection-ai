```python
# importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
# Load the dataset
df = pd.read_csv("../data/raw/creditcard.csv")
df.head()
```

```python
# Data description
df.shape
df.info()
df.describe()
```

## Basic Data Understanding

### Number of Transactions
- The dataset contains **284,807 financial transactions**, representing real-world credit card activity.

### Number of Features
- The dataset has **31 columns** in total.
- **30 feature columns** (`Time`, `V1`–`V28`, `Amount`)
- **1 target column** (`Class`), where:
  - `0` = Non-Fraudulent transaction  
  - `1` = Fraudulent transaction

### Data Types
- **30 columns** are of type `float64`, representing numerical transaction features.
- **1 column** (`Class`) is of type `int64`, representing the binary classification label.

### Missing Values
- There are **no missing values** in the dataset.
- All **284,807 rows are complete**, which simplifies preprocessing and model training.

### Initial Observations
- Most features (`V1`–`V28`) are anonymized and scaled, indicating prior transformation for privacy protection.
- The `Amount` feature shows a wide range of transaction values, suggesting the presence of both small and high-value transactions.
- The target variable (`Class`) has a very low mean value (~0.17%), indicating **severe class imbalance**, a common and critical challenge in fraud detection systems.


```python
# Checking for null values and missing values
df.isnull().sum()
# df.isnull().sum().sum()
# Percentage of missing values
(df.isnull().sum() / len(df)) * 100

```

## Missing Values Analysis

The dataset was examined for missing values across all features. 
No missing values were detected, indicating that the dataset is complete 
and suitable for downstream preprocessing and model training without 
the need for imputation.


```python
# Fraud vs Non-Fraud Distribution
fraud_counts = df["Class"].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values)
plt.xticks([0,1], ["Non-Fraud", "Fraud"])
plt.title("Fraud vs Non-Fraud Transactions")
plt.ylabel("Count")
plt.xlabel("Transaction Type")
plt.show()
```

```python
# Saving the plot
plt.savefig("../results/fraud_distribution.png")
```

```python
# Percentage of Fraud Transactions
fraud_percentage = fraud_counts[1] / fraud_counts.sum() * 100
fraud_percentage
```

Fraud accounts for less than 1% of transactions, highlighting the challenge of detecting rare but high-impact events.

```python
# Feature Distribution
plt.figure(figsize=(10,4))
sns.histplot(df["Amount"], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
# Saving the histogram
plt.savefig("../results/amount_distribution.png")
```
