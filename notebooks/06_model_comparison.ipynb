```python
# Model comparison
import pandas as pd
import matplotlib.pyplot as plt
roc_auc     = 0.9720834996210077
roc_auc_rf  = 0.9804269836803601
roc_auc_xgb = 0.9801210263227407

rec_lr  = 0.9183673469387755
rec_rf  = 0.8163265306122449
rec_xgb = 0.8367346938775511


results = pd.DataFrame({
    "Model": ["Logistic Regression", "Random Forest", "XGBoost"],
    "ROC_AUC": [roc_auc, roc_auc_rf, roc_auc_xgb],
    "Fraud_Recall": [rec_lr, rec_rf, rec_xgb]
})

results.round(4)
```

```python
# Visual Comparison
results.set_index("Model")[["ROC_AUC", "Fraud_Recall"]].plot(
    kind="bar", figsize=(8,4)
)

plt.title("Fraud Detection Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1)

plt.savefig("../results/model_comparison.png", bbox_inches="tight")
plt.show()

```

## Key Findings

- Logistic Regression provides a transparent baseline but struggles with complex fraud patterns.
- Random Forest improves detection by capturing non-linear relationships.
- XGBoost achieves the strongest overall performance, balancing fraud recall and false positives.
